```python
!pip install ultralytics -q
from ultralytics import YOLO
import os, yaml


model = YOLO("/content/yolo26n200.pt")
# Pastikan data.yaml sudah ada dan path test-nya benar
metrics = model.val(
    data   = yaml_path,   # path ke data.yaml
    split  = "test",      # gunakan test set, bukan val
    imgsz  = 640,
    batch  = 16,
    device = 'cpu',           # atau 'cpu'
    save_json = True,     # simpan hasil dalam format COCO JSON
    plots  = True,        # generate confusion matrix, PR curve, dll
)

p = metrics.box.mp
r = metrics.box.mr
f1_score = 2 * (p * r) / (p + r) if (p + r) > 0 else 0.0

print(f"mAP50     : {metrics.box.map50:.4f}")
print(f"mAP50-95  : {metrics.box.map:.4f}")
print(f"Precision : {metrics.box.mp:.4f}")
print(f"Recall    : {metrics.box.mr:.4f}")
print(f"F1 Score  : {f1_score:.4f}")
```
