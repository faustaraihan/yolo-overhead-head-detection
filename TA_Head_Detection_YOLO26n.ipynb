```python
!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="")
project = rf.workspace("bus-passenger-object-detection").project("overhead-passenger-detection-aacjr")
version = project.version(2)
dataset = version.download("yolo26")
```

```python
import os
import yaml

dataset_path = dataset.location
print(f"📂 Root: {dataset_path}\n")

# Lihat isi folder
for split in ["train", "valid", "test"]:
    img_path = os.path.join(dataset_path, split, "images")
    if os.path.exists(img_path):
        count = len(os.listdir(img_path))
        print(f"  [{split.upper()}] {count} gambar")

# Baca dan tampilkan data.yaml
yaml_path = os.path.join(dataset_path, "data.yaml")
with open(yaml_path, "r") as f:
    data_config = yaml.safe_load(f)

print(f"\n📋 data.yaml:")
print(f"  Classes   : {data_config['names']}")
print(f"  Num class : {data_config['nc']}")
print(f"  Train path: {data_config.get('train')}")
print(f"  Val path  : {data_config.get('val')}")
print(f"  Test path : {data_config.get('test')}")
```

```python
import yaml

yaml_path = os.path.join(dataset.location, "data.yaml")

with open(yaml_path, "r") as f:
    data_config = yaml.safe_load(f)

# Update path ke absolut agar tidak error saat training
data_config["train"] = os.path.join(dataset.location, "train", "images")
data_config["val"]   = os.path.join(dataset.location, "valid", "images")
data_config["test"]  = os.path.join(dataset.location, "test", "images")

with open(yaml_path, "w") as f:
    yaml.dump(data_config, f)

print("✅ data.yaml berhasil diupdate!")
print(yaml.dump(data_config, default_flow_style=False))
```

```python
from ultralytics import YOLO

# Load model YOLO26 pretrained (nano = paling ringan, cocok untuk Colab)
# Pilihan ukuran: yolo26n, yolo26s, yolo26m, yolo26l, yolo26x
model = YOLO("yolo26n.pt")

# Training
results = model.train(
    data      = yaml_path,        # Path ke data.yaml
    epochs    = 100,              # Jumlah epoch
    imgsz     = 640,              # Ukuran gambar
    batch     = 16,               # Batch size (turunkan ke 8 jika OOM)
    patience  = 20,               # Early stopping
    device    = 0,                # GPU (ganti ke 'cpu' jika tidak ada GPU)
    project   = "overhead_head_detection",
    name      = "yolo26_passenger",
    exist_ok  = True,

    # Augmentasi (cocok untuk overhead view)
    flipud    = 0.5,              # Flip vertikal (relevan untuk top-down)
    fliplr    = 0.5,              # Flip horizontal
    degrees   = 45,               # Rotasi (kepala bisa menghadap segala arah)
    scale     = 0.5,              # Scale augmentation
    mosaic    = 1.0,              # Mosaic augmentation
    hsv_h     = 0.015,
    hsv_s     = 0.7,
    hsv_v     = 0.4,

    # Optimizer
    optimizer = "AdamW",
    lr0       = 0.001,
    lrf       = 0.01,
    weight_decay = 0.0005,
)

print("✅ Training selesai!")
```
